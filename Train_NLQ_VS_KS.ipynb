## Generating Realistic E-commerce Queries: A Deep Dive

This post explores the code behind a Python script designed to generate realistic e-commerce search queries.  This is crucial for training search ML models, training search algorithms, and understanding user search behavior. The script generates two main types of queries: Keyword Searches (KS) and Natural Language Queries (NLQ).  We'll break down the logic behind each type and highlight the key features that make these queries realistic.

### Understanding the Data Sources

Before diving into query generation, let's look at the ingredients.  The script uses several lists and dictionaries to simulate the variety of products, brands, features, and user intents found in a real e-commerce environment. These include:

*   **Electronics:**  A comprehensive list of electronic product types (e.g., "TV", "smartphone", "laptop").
*   **Brands:**  A list of popular electronics brands (e.g., "Samsung", "Apple", "Sony").
*   **Models:**  A dictionary mapping brands to specific product models (e.g., "Samsung": ["Galaxy S23", "QLED TV"]).
*   **Samsung Specific Data**: These lists include the parameters that are used to make Samsung related searches more detailed. These include data related to ecosystem, accessories and product categories.
*   **Features:**  A list of common product features (e.g., "4K", "noise cancelling", "waterproof").
*   **Attributes:**  A list of product attributes users might search for (e.g., "screen size", "battery life", "price").
*   **Scenarios:**  A list of user contexts or needs (e.g., "for gaming", "for photography", "on a budget").
*   **Problems:**  A list of common issues users might experience with their devices (e.g., "cracked screen", "battery draining fast").
* **Comparisons, Negations, Currencies, and Idioms**: For expanding the queries in a more natural manner.
*   **Conditions:**  A list of product conditions (e.g., "refurbished", "used", "new").
*   **Typos:**  A dictionary of common misspellings for brand names, adding a touch of realism.

These data sources provide the building blocks for constructing a wide range of search queries.

### Keyword Search (KS) Generation

Keyword searches are typically shorter and consist of a combination of terms that users might type directly into a search bar.  They often lack the grammatical structure of a complete sentence.  The script generates KS queries by randomly selecting and combining elements from the data sources mentioned above.  Here's a breakdown of the logic:

1.  **Prioritize Brand and Model:**  With a high probability, the script starts by including a brand and, potentially, a specific model. This mimics users who know exactly what they're looking for. It also has capability to enter partial model name.

2.  **Add Electronic Type:**  The script often includes the type of electronic product, further refining the search.

3.  **Incorporate Features:**  Randomly selected features are added to represent users specifying desired characteristics.

4.  **Include Specs and Numbers:**  The script adds numerical specifications like screen size, storage capacity, or price ranges, mimicking users who have specific requirements.

5.  **Inject Typos:**  To simulate real-world user input, the script introduces common misspellings of brand names.

6.  **Refurbished/Used Queries:** Generates keywords related to refurbished products.

7.  **Attribute Focused KS**: These queries focus on attributes like best or good, followed by an attribute.

8. **Cheap/Affordable KS**: These queries target price sensitive users.

9. **Special characters**: This incorporates special characters like `/` and `-` in the queries.

The result is a set of keyword-based queries that resemble those seen in real e-commerce search logs.

### Natural Language Query (NLQ) Generation

Natural Language Queries, on the other hand, are phrased as complete sentences or questions, reflecting how users might naturally express their needs.  The script uses a template-based approach to generate NLQs, combined with random selection from the data sources.

1.  **Template Selection:**  The script starts by randomly choosing from a large set of pre-defined NLQ templates.  These templates cover a wide range of query types, including:
    *   Product recommendations ("What's the best {electronic} for {scenario}?")
    *   Specific feature requests ("I need a {electronic} with {feature}.")
    *   Brand-specific queries ("Can you recommend a {brand} {electronic}?")
    *   Comparisons ("Which {electronic} is better, {brand} {model_1} or {brand} {model_2}?")
    *   Problem-solving ("My {brand} {electronic} has a {problem}, what are my options?")
    *   Price-constrained searches ("What are the top-rated {brand} {electronic} under {currency}{price}?")
    *   Attribute-focused queries ("I want a {brand} {electronic} with a really good {attribute}.")
    *   Samsung-specific queries. ("How do I use {feature} on my {samsung_model}?")
    *   ... and many more variations.

2.  **Parameter Filling:**  Once a template is chosen, the script fills in the placeholders (e.g., `{electronic}`, `{scenario}`, `{feature}`) with randomly selected values from the data sources.  This ensures variety and covers a broad spectrum of possible queries.
    *   The script carefully handles comparisons, ensuring that different brands and models are selected when required by the template.

3.  **Contradictory Features**: Adds queries to find products with contradictory features.

4.  **Compatibility**: Adds queries to check for compatibility between different brands.

5.  **Grammatical Imperfections**: To mimic human error, the generated queries might intentionally omit articles ("a", "the") or have incorrect singular/plural forms.

6. **Natural Language Variations:** The code incorporates subtle changes to wording, like replacing "What's" with "What is" or "I'm" with "I am", further enhancing the natural feel of the queries.

7. **Adding pleasantries:** Adds variations of pleasantries and words like `thanks` and `please help`.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support, accuracy_score, confusion_matrix
from transformers import (
    AutoTokenizer, AutoModelForSequenceClassification,
    Trainer, TrainingArguments
)
import torch
from datasets import Dataset
import time
from tabulate import tabulate
import seaborn as sns
import matplotlib.pyplot as plt

class QueryClassifier:
    def __init__(self, model_name):
        self.model_name = model_name
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = None
        self.label2id = {'KS': 0, 'NLQ': 1}
        self.id2label = {0: 'KS', 1: 'NLQ'}
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        print(f"Using device: {self.device}")

    def prepare_data(self, df):
        """Prepare dataset for training"""
        # Convert labels to numeric
        df['label'] = df['label'].map(self.label2id)

        # Split data
        train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42)
        val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

        # Convert to HuggingFace datasets
        train_dataset = Dataset.from_pandas(train_df)
        val_dataset = Dataset.from_pandas(val_df)
        test_dataset = Dataset.from_pandas(test_df)

        # Tokenize all datasets
        train_dataset = train_dataset.map(
            lambda x: self.tokenize_function(x['query']),
            remove_columns=train_dataset.column_names,
            batched=True
        )
        val_dataset = val_dataset.map(
            lambda x: self.tokenize_function(x['query']),
            remove_columns=val_dataset.column_names,
            batched=True
        )
        test_dataset = test_dataset.map(
            lambda x: self.tokenize_function(x['query']),
            remove_columns=test_dataset.column_names,
            batched=True
        )

        # Add labels back to datasets
        train_dataset = train_dataset.add_column('labels', train_df['label'].tolist())
        val_dataset = val_dataset.add_column('labels', val_df['label'].tolist())
        test_dataset = test_dataset.add_column('labels', test_df['label'].tolist())

        return train_dataset, val_dataset, test_dataset

    def tokenize_function(self, examples):
        """Tokenize text data"""
        return self.tokenizer(
            examples,
            padding='max_length',
            truncation=True,
            max_length=128,
            return_tensors=None  # Return PyTorch tensors
        )

    def train(self, train_dataset, val_dataset):
        """Train the model"""
        # Initialize model and move to correct device
        self.model = AutoModelForSequenceClassification.from_pretrained(
            self.model_name,
            num_labels=2,
            label2id=self.label2id,
            id2label=self.id2label
        ).to(self.device)

        # Training arguments
        training_args = TrainingArguments(
            output_dir=f"./results_{self.model_name}",
            num_train_epochs=3,
            per_device_train_batch_size=16,
            per_device_eval_batch_size=16,
            evaluation_strategy="epoch",
            save_strategy="epoch",
            load_best_model_at_end=True,
            push_to_hub=False,
            no_cuda=not torch.cuda.is_available(),
        )

        # Initialize trainer
        trainer = Trainer(
            model=self.model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset,
        )

        # Train the model
        trainer.train()

        return trainer

    def evaluate(self, test_dataset):
        """Evaluate the model"""
        trainer = Trainer(model=self.model)
        predictions = trainer.predict(test_dataset)
        preds = np.argmax(predictions.predictions, axis=1)
        labels = test_dataset['labels']

        # Calculate metrics
        precision, recall, f1, _ = precision_recall_fscore_support(
            labels,
            preds,
            average='binary',
            pos_label=1
        )
        accuracy = accuracy_score(labels, preds)

        # Create confusion matrix
        cm = confusion_matrix(labels, preds)

        return {
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'accuracy': accuracy,
            'confusion_matrix': cm
        }

    def inference_time_test(self, test_queries):
        """Test inference time on sample queries"""
        results = []

        for query, true_label in test_queries:
            # Measure inference time
            start_time = time.time()

            # Tokenize and move to correct device
            inputs = self.tokenizer(
                query,
                padding='max_length',
                truncation=True,
                max_length=128,
                return_tensors="pt"
            ).to(self.device)

            # Predict
            with torch.no_grad():
                outputs = self.model(**inputs)
                prediction = torch.argmax(outputs.logits, dim=1).item()

            end_time = time.time()
            inference_time = end_time - start_time

            results.append({
                'query': query,
                'true_label': self.id2label[true_label],
                'predicted_label': self.id2label[prediction],
                'inference_time': inference_time
            })

        return results

def plot_confusion_matrix(cm, model_name):
    """Plot confusion matrix"""
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

def main():
    # Sample data
    sample_data = pd.DataFrame({
        'query': [
            'Logitech keyboard used',
            'I need microwave with dustproof.',
            'show me all nikon earbuds models.',
            'my samsung gaming console has a cracked screen...',
            'headphones/44/inch',
            'Razer 360Hz sampling rate',
            'Dell XPS 15',
            "I'm not sure which Dell electric toothbrush to...",
            'What is the most connectivity LG charger?',
            'how to use qled tv with other samsung devices?...'
        ],
        'label': ['KS', 'NLQ', 'NLQ', 'NLQ', 'KS', 'KS', 'KS', 'NLQ', 'NLQ', 'NLQ']
    })

    # Models to test
    models = [
        'bert-base-uncased',
        'distilbert-base-uncased',
        'distilroberta-base'
    ]

    results = {}
    inference_times = {}

    for model_name in models:
        print(f"\nTraining and evaluating {model_name}...")

        try:
            # Initialize classifier
            classifier = QueryClassifier(model_name)

            # Prepare data
            train_dataset, val_dataset, test_dataset = classifier.prepare_data(sample_data)

            # Train model
            trainer = classifier.train(train_dataset, val_dataset)

            # Evaluate model
            metrics = classifier.evaluate(test_dataset)
            results[model_name] = metrics

            # Plot confusion matrix
            plot_confusion_matrix(metrics['confusion_matrix'], model_name)

            # Test inference time
            test_queries = list(zip(
                sample_data['query'][:10],
                [classifier.label2id[label] for label in sample_data['label'][:10]]
            ))
            inference_results = classifier.inference_time_test(test_queries)
            inference_times[model_name] = inference_results

            # Clear GPU memory
            if torch.cuda.is_available():
                torch.cuda.empty_cache()

        except Exception as e:
            print(f"Error processing {model_name}: {str(e)}")
            continue

    # Print results
    if results:
        print("\nModel Performance Comparison:")
        comparison_table = []
        for model_name, metrics in results.items():
            comparison_table.append([
                model_name,
                f"{metrics['precision']:.3f}",
                f"{metrics['recall']:.3f}",
                f"{metrics['f1']:.3f}",
                f"{metrics['accuracy']:.3f}"
            ])

        print(tabulate(
            comparison_table,
            headers=['Model', 'Precision', 'Recall', 'F1-Score', 'Accuracy'],
            tablefmt='grid'
        ))

    if inference_times:
        print("\nInference Time Comparison:")
        inference_comparison = []
        for model_name, results in inference_times.items():
            avg_time = np.mean([r['inference_time'] for r in results])
            inference_comparison.append([
                model_name,
                f"{avg_time:.4f}"
            ])

        print(tabulate(
            inference_comparison,
            headers=['Model', 'Average Inference Time (s)'],
            tablefmt='grid'
        ))

if __name__ == "__main__":
    main()
```
