**Summary of the Code**

The Jupyter Notebook focuses on generating and analyzing spelling variations (typos) for a predefined list of Samsung product names. It achieves this through a combination of techniques:

1.  **Typo Generation:**
    *   **Keyboard Adjacency:** Simulates typos based on keys that are physically close to each other on a QWERTY keyboard (e.g., "s" becoming "a" or "d").
    *   **Character Swapping:** Swaps adjacent characters (e.g., "galaxy" becoming "aglaxy").
    *   **Missing Characters:** Removes characters (e.g., "samsung" becoming "amsun").
    *   **Spacing Variations:**  Handles variations in spacing (e.g., "samsung galaxy" vs. "samsunggalaxy").
    *   **Abbreviations:** Uses a predefined list of common Samsung abbreviations (e.g., "Samsung" -> "SS", "Galaxy" -> "G").
    *   **Model-Only:** Generates variants consisting of only the model name (e.g., "Galaxy S25 Ultra" -> "S25 Ultra").
    * **Combination:** Generates typos by combining any of the methods above

    *Example:*  For the product "Samsung Galaxy S25 Ultra", the code might generate typos like "samsugn galxy s25 ultra", "galaxy s25", "SSGS25Ultra", "Samsng Glaxy S25 Ultr", etc.

````python
# ## Typo Generation Functions

# These functions are the core of the typo generation process.  They create variations of the product names based on different typo strategies.

import random
import re

def generate_variants_with_columns(product):
    """
    Generates typo variants for a given product name, tracking the original term used for the variant.

    Args:
        product (str): The original product name.

    Returns:
        list: A list of lists, where each inner list contains:
            - The variant (typo).
            - The original product name.
            - The original term that was modified to create the variant.
    """
    words = product.lower().split()
    variants = []

    # Basic spacing variations
    for variant in [' '.join(words), ''.join(words), '-'.join(words), '_'.join(words), '/'.join(words)]:
        variants.append([variant, product, product])

    # Samsung abbreviations (defined below)
    product_lower = product.lower()
    for term, abbr_list in samsung_abbreviations.items():
        if term.lower() in product_lower:
            for abbr in abbr_list:
                variant = product_lower.replace(term.lower(), abbr.lower())
                variants.append([variant, product, term])

                # Also try with no spaces
                no_space_variant = variant.replace(' ', '')
                variants.append([no_space_variant, product, term])

    # Model-only variants (remove brand names)
    if len(words) > 1:
        model_only = ' '.join(words[1:])
        variants.append([model_only, product, words[0]])
        variants.append([model_only.replace(' ',''), product, words[0]])

    # Realistic human-made typos (using helper function below)
    for word in words:
        word_typos = add_typo(word)
        for typo, original in word_typos.items():
            # Replace the original word with the typo in the full product name
            for i, w in enumerate(words):
                if w == word:
                    new_words = words.copy()
                    new_words[i] = typo
                    variant = ' '.join(new_words)
                    variants.append([variant, product, word])  # Track the specific word with the typo

    return variants



def add_typo(word):
    """
    Generates typos for a single word based on:
        - Missing characters.
        - Swapped adjacent characters.
        - Keyboard-adjacent replacements.

    Args:
        word (str): The word to generate typos for.

    Returns:
        dict: A dictionary where keys are typo variants and values are the original word.
    """
    typos = {}

    # Missing character typos (only for words longer than 3 characters)
    if len(word) > 3:
        idx = random.randint(0, len(word) - 1)
        typo = word[:idx] + word[idx+1:]
        typos[typo] = word

    # Swapped adjacent characters (only for words longer than 2 characters)
    if len(word) > 2:
        idx = random.randint(0, len(word) - 2)
        swapped = list(word)
        swapped[idx], swapped[idx+1] = swapped[idx+1], swapped[idx]
        typos[''.join(swapped)] = word

    # Keyboard-adjacent replacements
    for i, char in enumerate(word):
        if char in keyboard_adjacent:
            for typo_char in keyboard_adjacent[char]:
                typos[word[:i] + typo_char + word[i+1:]] = word

    return typos


# Common Samsung abbreviations and short forms
samsung_abbreviations = {
    # Brand and product line
    "Samsung": ["SS", "Sammy", "Sams", "SM", "SMSNG", "Sung"],
    "Galaxy": ["G", "Gal", "Glxy", "GX", "GLX"],
    "Samsung Galaxy": ["SS G", "Sammy G", "SG", "S Gal", "SGX"],

    # Device types
    "Smartphone": ["SP", "Phone", "Phn", "SmPh", "SPh"],
    "Tablet": ["Tab", "TB", "Tbl", "T", "Tbt"],
    "Watch": ["W", "Wtch", "WCH", "GW", "SW"],
    "Buds": ["B", "Bds", "Earbuds", "EB", "GB"],

    # Series identifiers
    "Series": ["S", "Sr", "Ser", "Srs"],
    "Ultra": ["U", "Ult", "ULT", "Ultr"],
    "Plus": ["+", "Pl", "Pls", "P"],
    "Pro": ["P", "Pr", "Prof"],
    "Fan Edition": ["FE", "F.E.", "Fan Ed", "F Ed"],
    "Fold": ["F", "Fld", "FD", "Folding"],
    "Flip": ["FL", "Flp", "FLP", "F"],
    "Active": ["A", "Act", "Actv", "ATV"],
    "Classic": ["C", "Cls", "Clsc", "CL"],

    # Appliances
    "Bespoke": ["BSPK", "Bsp", "Besp", "BSP", "BPK"],
    "Refrigerator": ["Fridge", "Ref", "Rfg", "Refr", "FR"],
    "Washer": ["Wshr", "WS", "WSH", "W", "Wsh"],
    "Dryer": ["Dryr", "DR", "DRY", "D", "Dry"],
    "Microwave": ["Micro", "MW", "Mic", "MWO", "MV"],
    "Induction": ["Indctn", "Ind", "IND", "Indc", "IC"],
    "Dishwasher": ["Dshwshr", "DW", "DSH", "DWash", "DishW"],

    # TV and display
    "QLED": ["Q", "QL", "QLD", "Q-LED"],
    "OLED": ["O", "OL", "OLD", "O-LED"],
    "Neo QLED": ["NQ", "Neo Q", "NQL", "Neo"],
    "The Frame": ["Frame", "TF", "Fr", "FRM"],
    "The Serif": ["Serif", "TS", "SR", "SRF"],
    "The Terrace": ["Terrace", "TT", "TR", "TRC"],
    "Monitor": ["Mntr", "Mon", "M", "MT", "MON"],

    # Other common terms
    "Wireless": ["Wrls", "W", "WL", "Wrlss", "Wire"],
    "Bluetooth": ["BT", "BTH", "Blue", "B-T"],
    "Smart": ["Smt", "SM", "S", "Smrt"],
    "Premium": ["Prem", "PR", "PRM", "Prm"]
}
````

````python
# ## Generate and Save Initial DataFrame

# This cell generates the initial DataFrame containing all variants, original product names, and the original terms that were modified.  It then saves this DataFrame to a pickle file for later use.  This avoids regenerating the data every time the notebook is run.

import pandas as pd
import pickle

# Generate expanded list with columnar structure
expanded_list = []
for name in product_names:
    variants = generate_variants_with_columns(name)
    expanded_list.extend(variants)

# Convert to DataFrame
columns = ["Variant", "Original Name", "Original Term"]
df = pd.DataFrame(expanded_list, columns=columns)

# Save the DataFrame to a pickle file
with open('expanded_samsung_product_dict_with_terms.pkl', 'wb') as f:
    pickle.dump(df, f)

print("New dictionary saved as 'expanded_samsung_product_dict_with_terms.pkl'")
print("\nFirst 10 rows of the updated DataFrame:")
print(df.head(10))
````

````python
# ## Create Grouped Dictionary

# This cell loads the DataFrame created in the previous step and transforms it into a dictionary. This dictionary, `grouped_dict_with_terms`, is the primary data structure used for most of the subsequent operations.  It groups variants by the *variant* itself and stores a list of associated original product names and original terms. This structure is more efficient for lookups and analysis than the flat DataFrame.

import pickle
import pandas as pd
from collections import defaultdict

# Load the existing expanded dictionary (now a DataFrame)
try:
    with open('expanded_samsung_product_dict_with_terms.pkl', 'rb') as f:
        df = pickle.load(f)
    print(f"Loaded existing DataFrame with {len(df)} variants")
except FileNotFoundError:
    print("Error: expanded_samsung_product_dict_with_terms.pkl not found!")
    exit(1)


# Create a more detailed grouped dictionary that includes original terms
grouped_dict_with_terms = {}

for _, row in df.iterrows():
    variant = row['Variant']
    original_name = row['Original Name']
    original_term = row['Original Term']

    if variant not in grouped_dict_with_terms:
        grouped_dict_with_terms[variant] = {
            'original_names': [],
            'original_terms': []
        }

    grouped_dict_with_terms[variant]['original_names'].append(original_name)
    grouped_dict_with_terms[variant]['original_terms'].append(original_term)

# Remove duplicates
for variant in grouped_dict_with_terms:
    grouped_dict_with_terms[variant]['original_names'] = list(set(grouped_dict_with_terms[variant]['original_names']))
    grouped_dict_with_terms[variant]['original_terms'] = list(set(grouped_dict_with_terms[variant]['original_terms']))

# Save the enhanced grouped dictionary
with open('grouped_samsung_product_dict_with_terms.pkl', 'wb') as f:
    pickle.dump(grouped_dict_with_terms, f)

print(f"Enhanced grouped dictionary saved as 'grouped_samsung_product_dict_with_terms.pkl'")
````

````python
# ## Visualization Functions

# These functions provide different ways to visualize the generated data, helping to understand the distribution and nature of the typos.

import matplotlib.pyplot as plt
import networkx as nx
from wordcloud import WordCloud
from tabulate import tabulate
from rapidfuzz.distance import DamerauLevenshtein
import random

def visualize_variants(product_name, grouped_dict):
    """
    Visualizes all generated variants for a specific product name,
    including the original terms used to create them.

    Args:
        product_name (str): The product name to visualize variants for.
        grouped_dict (dict): The grouped dictionary of variants.
    """
    print(f"Generated variants for: {product_name}")
    count = 0

    # Find all variants that map to this product
    for variant, info in grouped_dict.items():
        if product_name in info['original_names']:
            count += 1
            original_terms = info['original_terms']
            terms_str = ", ".join(original_terms)
            print(f"{count}. {variant} (Terms: {terms_str})")

    if count == 0:
        print("No variants found for this product.")
        return False
    else:
        print(f"Total variants: {count}")
        return True

def visualize_grouped_keys(key, grouped_dict):
    """
    Visualizes the products and original terms associated with a specific variant key.

    Args:
        key (str): The variant key to visualize.
        grouped_dict (dict): The grouped dictionary of variants.
    """
    print(f"Products associated with '{key}':")
    if key in grouped_dict:
        info = grouped_dict[key]
        print("\nOriginal Products:")
        for idx, product in enumerate(info['original_names'], start=1):
            print(f"{idx}. {product}")

        print("\nOriginal Terms:")
        for idx, term in enumerate(info['original_terms'], start=1):
            print(f"{idx}. {term}")
    else:
        print("No products found for this key.")

        # Try to find similar keys
        similar_keys = [k for k in grouped_dict.keys() if key.lower() in k.lower()]
        if similar_keys:
            print("\nSimilar keys found:")
            for idx, similar_key in enumerate(similar_keys[:5], start=1):
                print(f"{idx}. {similar_key}")
            if len(similar_keys) > 5:
                print(f"...and {len(similar_keys) - 5} more")


def plot_variant_counts(df, selected_products, title="Variant Counts for Selected Samsung Products"):
    """
    Plots a bar chart of the variant counts for a selection of products.

    Args:
        df (pd.DataFrame): The DataFrame containing variant and product information.  Must have 'Original Name' column
        selected_products (list): A list of product names to include in the plot.
        title (str, optional): The title of the plot. Defaults to "Variant Counts...".
    """
    # Filter the DataFrame for the selected products
    filtered_df = df[df['Original Name'].isin(selected_products)]

    # Count occurrences of each selected product's variants
    variant_counts = filtered_df['Original Name'].value_counts()

    # Plot settings
    plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
    bars = plt.bar(variant_counts.index, variant_counts.values, color='skyblue')
    plt.title(title, fontsize=16)
    plt.xlabel("Product Name", fontsize=12)
    plt.ylabel("Number of Variants", fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels

    # Add counts on top of the bars
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 1, str(int(bar.get_height())), ha='center', fontsize=10)

    plt.tight_layout()  # Adjust layout to prevent label cutoff
    plt.show()

def plot_top_variant_counts(df, top_n=20, title="Top 20 Original Names with Their Variant Counts"):
    """
    Plots the top N products with the highest number of variants.

    Args:
        df (pd.DataFrame): The DataFrame containing variant and product information.
        top_n (int): number of products to be displayed
        title (str): title of the plot
    """
    variant_counts = df['Original Name'].value_counts()

    plt.figure(figsize=(15, 8))
    variant_counts.head(top_n).plot(kind='bar', color='skyblue')
    plt.title(title, fontsize=16)
    plt.xlabel("Original Name", fontsize=12)
    plt.ylabel("Number of Variants", fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.tight_layout()
    plt.show()

def plot_term_distribution(df, selected_products, top_n=15):
    """
    Visualizes the distribution of original terms for a selection of products.
    Args:
        df (pd.DataFrame): The DataFrame containing variant, product, and term information.
                           Must include the 'Original Term' column.
        selected_products (list): A list of product names to filter the data.
        top_n (int): show the top N terms
    """
    filtered_df = df[df['Original Name'].isin(selected_products)]

    plt.figure(figsize=(12, 6))
    term_counts = filtered_df['Original Term'].value_counts().head(top_n)
    term_bars = plt.bar(term_counts.index, term_counts.values, color='lightgreen')
    plt.title("Distribution of Original Terms for Selected Products", fontsize=16)
    plt.xlabel("Original Term", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)

    for bar in term_bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 1, str(int(bar.get_height())), ha='center', fontsize=10)

    plt.tight_layout()
    plt.show()

def create_network_diagram(product_name, alternate_names, grouped_dict):
    """
    Creates a network diagram showing the relationships between a product and its variants.

    Args:
        product_name (str): The original product name.
        alternate_names (list): A list of variant names.
        grouped_dict (dict): The grouped dictionary of variants.
    """
    G = nx.Graph()
    G.add_node(product_name, color='green', size=1000)  # Original product name node

    # Group variants by original term
    term_groups = {}
    for name in alternate_names:
        terms = grouped_dict[name]['original_terms']
        term = terms[0] if terms else "Unknown"
        if term not in term_groups:
            term_groups[term] = []
        term_groups[term].append(name)

    # Add term nodes as intermediaries
    for term, variants in term_groups.items():
        G.add_node(term, color='red', size=700)
        G.add_edge(product_name, term)

        for name in variants[:5]:  # Limit to 5 variants per term to avoid overcrowding
            G.add_node(name, color='blue', size=300)
            G.add_edge(term, name)

        if len(variants) > 5:
            more_node = f"{term} (+{len(variants)-5} more)"
            G.add_node(more_node, color='yellow', size=400)
            G.add_edge(term, more_node)

    pos = nx.spring_layout(G, k=0.3, iterations=50)  # Adjust layout parameters

    # Extract node attributes
    colors = [data['color'] for _, data in G.nodes(data=True)]
    sizes = [data['size'] for _, data in G.nodes(data=True)]

    plt.figure(figsize=(14, 10))
    nx.draw(G, pos, with_labels=True, node_color=colors, node_size=sizes,
            font_size=8, font_weight='bold', edge_color='gray', width=0.5,
            alpha=0.8)
    plt.title(f"Network Diagram of Variants for '{product_name}'")
    plt.tight_layout()
    plt.show()

def create_word_cloud(alternate_names):
    """
    Generates a word cloud from a list of variant names.

    Args:
        alternate_names (list): A list of variant names.
    """
    text = " ".join(alternate_names)
    wordcloud = WordCloud(width=800, height=400, background_color="white",
                         max_words=100, collocations=False).generate(text)

    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title("Word Cloud of Generated Variants")
    plt.show()

def generate_table(dictionary, rows=10, show_high_distances=True):
    """
     Generates a table showing variants, their closest original words, and Damerau-Levenshtein distances.
    Args:
        dictionary: grouped dictionary
        rows: number of rows
        show_high_distances: display by high distances

    Returns: table

    """
    table_data = []

    for variant, originals in dictionary.items():
        if originals:  # Make sure the list isn't empty
            original = originals[0]  # Take the first product name from the list
            closest_word, distance = find_closest_word(variant, original)
            table_data.append([variant, original, closest_word, distance])

    # Sort by distance (high to low if show_high_distances is True)
    if show_high_distances:
        table_data = sorted(table_data, key=lambda x: x[3], reverse=True)[:rows]
    else:
        table_data = sorted(table_data, key=lambda x: x[3])[:rows]

    # Create a tabular representation using tabulate
    headers = ["Newly Generated Word", "Actual Correct Word", "Closest Word", "Damerau-Levenshtein Distance"]
    return tabulate(table_data, headers=headers, tablefmt="grid")

def find_closest_word(variant, original):
    """
      find the closest word in the original text.

    Args:
        variant: generated variant
        original: original word

    Returns: closest word and distance

    """
    original_words = original.lower().split()
    closest_word = None
    min_distance = float('inf')

    for word in original_words:
        distance = DamerauLevenshtein.distance(variant, word)
        if distance < min_distance:
            min_distance = distance
            closest_word = word

    return closest_word, min_distance

def get_alternate_names(product_name, expanded_dict):
    """
    Retrieves all alternate names (variants) for a given product name from the expanded dictionary.

    Args:
        product_name (str): The original product name.
        expanded_dict (dict): The dictionary mapping variants to original product information.

    Returns:
        list: A list of variant names.  Returns an empty list if no variants are found.
    """
    alternate_names = []
    for variant, info in expanded_dict.items():
        if product_name in info['original_names']:
            alternate_names.append(variant)
    return alternate_names

def calculate_distances(product_name, alternate_names):
    """
      calculate Damerau-Levenshtein distances for each variant.

    Args:
        product_name: product name
        alternate_names: generated variants

    Returns: distances

    """
    rows = []
    for variant in alternate_names:
        distance = DamerauLevenshtein.distance(variant, product_name.lower())
        # Get the original term that was modified
        original_term = grouped_dict_with_terms[variant]['original_terms'][0] if grouped_dict_with_terms[variant]['original_terms'] else "Unknown"
        rows.append([variant, product_name, original_term, distance])
    return rows
````

````python
# ## Fuzzy Matching and Correction

from rapidfuzz import process, fuzz

def correct_product_name(query, dictionary, threshold=80):
    """
    Corrects a misspelled product query using fuzzy matching.

    Args:
        query (str): The misspelled query.
        dictionary (dict): The dictionary of variants and original product names.
        threshold (int): The minimum similarity score (0-100) to consider a match.

    Returns:
        list: The original product name(s) associated with the best match, or None if no good match is found.
    """
    match = process.extractOne(query.lower(), list(dictionary.keys()), scorer=fuzz.WRatio)

    if match and match[1] >= threshold:
        return dictionary[match[0]]['original_names']  # Return original product name mapped to matched variant

    return None

# --- Example Usage of Fuzzy Matching ---
misspelled_examples = ["galxy s22 ultr", "glxybds", "s25ultra", "22ultr", "samaung fold","samsugn wathc", "bespoke refrigirator"]
for misspelled in misspelled_examples:
    corrected = correct_product_name(misspelled, grouped_dict_with_terms)
    print(f"Misspelled: {misspelled} -> Suggestion: {corrected}")
````

````python
# Cell 2: Product data and keyboard layout
# product_names = [...]  # (User's product list here)

# keyboard_adjacent = {...}  # (User's keyboard layout here)

# Normalization function
def normalize_product_name(name):
    """Standardize product names for consistent processing"""
    return re.sub(r'[^a-z0-9 ]', '', name.lower().strip())

# Precompute normalized names
normalized_products = {normalize_product_name(p): p for p in product_names}
````

3. Advanced Typo Generation


````python
# Cell 3: Typo generation engine
def generate_typo_variants(product):
    """Generate realistic typo variants with tracking"""
    base = normalize_product_name(product)
    variants = defaultdict(list)

    # Spacing variations
    for sep in ['', '-', '_', '/']:
        variant = sep.join(base.split())
        variants[variant].append(('spacing', product))

    # Keyboard substitution
    for i in range(len(base)):
        if base[i] in keyboard_adjacent:
            for sub in keyboard_adjacent[base[i]]:
                typo = base[:i] + sub + base[i+1:]
                variants[typo].append(('keyboard', product))

    # Common abbreviations
    abbreviation_map = {
        'galaxy': ['g', 'glx', 'gal'],
        'ultra': ['ult', 'u'],
        'plus': ['+', 'pls']
    }
    for term, abbrs in abbreviation_map.items():
        if term in base:
            for abbr in abbrs:
                variant = base.replace(term, abbr)
                variants[variant].append(('abbreviation', product))

    return variants


````

4. Optimized Dictionary Creation

````python
# Cell 4: Dictionary builder
def build_typo_dictionary(products):
    """Create optimized typo-to-product mapping"""
    master_dict = defaultdict(lambda: {'products': [], 'types': []})

    for product in products:
        variants = generate_typo_variants(product)
        for variant, sources in variants.items():
            for source_type, src_product in sources:
                if src_product not in master_dict[variant]['products']:
                    master_dict[variant]['products'].append(src_product)
                    master_dict[variant]['types'].append(source_type)

    # Add canonical forms
    for norm_name, orig_name in normalized_products.items():
        master_dict[norm_name]['products'].insert(0, orig_name)
        master_dict[norm_name]['types'].insert(0, 'canonical')

    return dict(master_dict)

# Build and save dictionary
typo_dict = build_typo_dictionary(product_names)
with open('samsung_typo_dict.pkl', 'wb') as f:
    pickle.dump(typo_dict, f)

````

````python
# Cell 6: Fuzzy matching correction (fixed)
from rapidfuzz.distance import DamerauLevenshtein  # Add this import

class ProductCorrector:
    def __init__(self, typo_dict):
        self.typo_dict = typo_dict
        self.all_variants = list(typo_dict.keys())

    def correct_query(self, query, threshold=85):
        """Correct user query with context-aware matching"""
        norm_query = normalize_product_name(query)

        # Exact match check
        if norm_query in self.typo_dict:
            return self._resolve_matches(norm_query)

        # Fuzzy match
        match = process.extractOne(
            norm_query,
            self.all_variants,
            scorer=fuzz.WRatio,
            score_cutoff=threshold
        )

        if match:
            return self._resolve_matches(match[0])
        return None

    def _resolve_matches(self, variant):
        """Resolve multiple product matches"""
        products = self.typo_dict[variant]['products']
        return {
            'variant': variant,
            'possible_products': products,
            'confidence': self._calculate_confidence(variant, products)
        }

    def _calculate_confidence(self, variant, products):
        """Calculate match confidence score"""
        base_score = 100 - DamerauLevenshtein.distance(variant, products[0].lower())
        return min(max(base_score - (len(products)-1)*5, 0), 100)
````

This ensures that your search system can intelligently resolve ambiguities and improve its ability to handle typos effectively.

Solution: Grouping and Prioritizing Keys

1. Group Keys by Product Names
Instead of mapping each key to a single product name, map the keys to a list of associated product names. This way, when a user searches for a typo, you can return all relevant products rather than penalizing the search for ambiguity.

2. Prioritize Results
If multiple products are associated with the same key, prioritize them based on relevance, popularity, or predefined rules (e.g., flagship models first).

````python
import pickle
from rapidfuzz.distance import DamerauLevenshtein
from tabulate import tabulate

# Load grouped dictionary with mappings
with open('grouped_samsung_product_dict.pkl', 'rb') as f:
    expanded_dict = pickle.load(f)

# Function to find the closest word in the original text
def find_closest_word(variant, original):
    original_words = original.lower().split()
    closest_word = None
    min_distance = float('inf')

    for word in original_words:
        distance = DamerauLevenshtein.distance(variant, word)
        if distance < min_distance:
            min_distance = distance
            closest_word = word

    return closest_word, min_distance

# Function to generate a table of variants with distances to closest words
def generate_table(dictionary, rows=10, show_high_distances=True):
    table_data = []

    for variant, originals in dictionary.items():
        if originals:  # Make sure the list isn't empty
            original = originals[0]  # Take the first product name from the list
            closest_word, distance = find_closest_word(variant, original)
            table_data.append([variant, original, closest_word, distance])

    # Sort by distance (high to low if show_high_distances is True)
    if show_high_distances:
        table_data = sorted(table_data, key=lambda x: x[3], reverse=True)[:rows]
    else:
        table_data = sorted(table_data, key=lambda x: x[3])[:rows]

    # Create a tabular representation using tabulate
    headers = ["Newly Generated Word", "Actual Correct Word", "Closest Word", "Damerau-Levenshtein Distance"]
    return tabulate(table_data, headers=headers, tablefmt="grid")

# Example usage: Display top 10 rows with highest distances
print(generate_table(expanded_dict, rows=10, show_high_distances=True))
````

# Handling Repetitive Keys and Typo Variants in Product Search

This document outlines a robust approach to handling typos and repetitive keys (e.g., "samsung/galaxy") in a product search system. The key improvement is to group typo variants and prioritize the associated product names, leading to a more resilient and user-friendly search experience.

## Problem: Repetitive Keys and Ambiguity

A naive typo handling system might map a typo variant (key) directly to a single product name. This creates problems when multiple products share similar names or have the same typo variants.  For example, "samsung/galaxy" could refer to phones, tablets, or watches.  Direct mapping penalizes searches and provides poor results.

## Solution: Grouping and Prioritizing Keys

The solution involves two main steps:

1.  **Grouping:** Map each typo variant (key) to a *list* of potential product names, rather than a single product.  This acknowledges the ambiguity.
2.  **Prioritizing:**  If a key maps to multiple products, prioritize the results based on relevance, popularity, or predefined rules (e.g., newer models first).

This approach allows the search system to return all relevant products, even with typos, and present them in a logical order.

```python
import pickle
import random
from collections import defaultdict

# Function to generate typo variants
def generate_variants(product):
    words = product.lower().split()
    variants = set()

    # Basic spacing variations
    variants.update([
        ' '.join(words),
        ''.join(words),
        '-'.join(words),
        '_'.join(words),
        '/'.join(words)
    ])

    # Abbreviations and short forms
    abbreviations = {
        'ultra': 'u',
        'galaxy': 'g',
        'samsung': 'ss',
        'bespoke': 'bspk',
        'refrigerator': 'fridge',
        'washer': 'wshr',
        'dryer': 'dryr',
        'microwave': 'micro',
        'induction': 'indctn',
        'dishwasher': 'dshwshr'
    }
    
    for word, abbr in abbreviations.items():
        if word in words:
            variants.add(product.lower().replace(word, abbr))
            variants.add(''.join(words).replace(word, abbr))

    # Model-only variants (remove brand names)
    if len(words) > 1:
        model_only = ' '.join(words[1:])
        variants.add(model_only)
        variants.add(model_only.replace(' ',''))

    # Realistic human-made typos:
    def add_typo(word):
        typos = set()
        
        # Missing character typos
        if len(word) > 3:
            idx = random.randint(0, len(word)-1)
            typos.add(word[:idx] + word[idx+1:])
        
        # Swapped adjacent characters
        if len(word) > 2:
            idx = random.randint(0, len(word)-2)
            swapped = list(word)
            swapped[idx], swapped[idx+1] = swapped[idx+1], swapped[idx]
            typos.add(''.join(swapped))
        
        # Keyboard-adjacent replacements
        for i, char in enumerate(word):
            if char in keyboard_adjacent:
                for typo_char in keyboard_adjacent[char]:  # Iterate through adjacent characters
                    typos.add(word[:i] + typo_char + word[i+1:])
        
        return typos
    
    for word in words:
        variants.update(add_typo(word))

    return list(variants)

# Generate grouped dictionary with mappings to original names
```

*   **`generate_variants(product)` Function:** This function takes a product name as input and generates a set of typo variants.  It includes:

    *   **Abbreviations:**  Creates short forms using common abbreviations (e.g., "ultra" to "u").
    *   **Model-only variants:** Extracts the model part of the name (e.g., "S25 Ultra" from "Galaxy S25 Ultra").
    *   **`add_typo(word)` Function:**  Simulates realistic typos:
        *   **Missing characters:**  Randomly removes a character.

### **Step 1: Generating Typo Variants**

The first step is to generate typo variants for product names. We use techniques like spacing variations, abbreviations, and realistic human-made typos (e.g., missing characters, swapped characters, and keyboard-adjacent replacements).

#### Code Example:
```python
import random

def generate_variants(product):
    words = product.lower().split()
    variants = set()

    # Basic spacing variations
    variants.update([
        ' '.join(words),
        ''.join(words),
        '-'.join(words),
        '_'.join(words),
        '/'.join(words)
    ])

    # Abbreviations and short forms
    abbreviations = {
        'ultra': 'u',
        'galaxy': 'g',
        'samsung': 'ss',
        'refrigerator': 'fridge',
        'washer': 'wshr',
        # Add more abbreviations as needed...
    }
    
    for word, abbr in abbreviations.items():
        if word in words:
            variants.add(product.lower().replace(word, abbr))

    # Realistic human-made typos
    def add_typo(word):
        typos = set()
        
        # Missing character typos
        if len(word) > 3:
            idx = random.randint(0, len(word)-1)
            typos.add(word[:idx] + word[idx+1:])
        
        # Swapped adjacent characters
        if len(word) > 2:
            idx = random.randint(0, len(word)-2)
            swapped = list(word)
            swapped[idx], swapped[idx+1] = swapped[idx+1], swapped[idx]
            typos.add(''.join(swapped))
        
        return typos
    
    for word in words:
        variants.update(add_typo(word))

    return list(variants)
```
